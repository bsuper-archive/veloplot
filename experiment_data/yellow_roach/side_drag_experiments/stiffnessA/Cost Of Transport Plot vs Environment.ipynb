```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
sns.set_palette("bright")
```

```python
figure, axarr = plt.subplots(1)

y = [100.86336624,62.2942299682,49.0995254557,38.0616476432]
x = [0.02555974843,0.04138492872,0.05250645995,0.06773333333]
n = [4,6,8,10]
axarr.plot(x,y,color='b',marker='o')
axarr.axis([0,0.1,0,120])
axarr.set_ylabel('Cost of Transport')
axarr.set_xlabel('Speed (m/s)')
axarr.text(0.06,110,u"\u2022 distance between flaps",fontsize=12,color='b')
for i, txt in enumerate(n):

    txt = str(txt)+'cm'
    axarr.annotate(txt, (x[i],y[i]+5),size=15,color='b')

plt.show()
```

```python
output_dir = "./out/"
output_filename = "COT_VS_SPEED.png"
if not os.path.exists(output_dir):
            os.makedirs(output_dir)
figure.savefig(output_dir + output_filename, dpi=450)
```

```python
figure, axarr = plt.subplots(1)

y = [100.86336624,62.2942299682,49.0995254557,38.0616476432]
x = [0.0451764467426,0.0185630645527,0.011004465099,0.00417047212162]

n = [4,6,8,10]
axarr.plot(x,y,color='b',marker='o')
axarr.axis([0,0.06,0,120])
axarr.set_ylabel('Cost of Transport')
axarr.set_xlabel('Drag Energy (Joules)')
axarr.text(0.036,112,u"\u2022 distance between flaps",fontsize=12,color='b')
for i, txt in enumerate(n):
    txt = str(txt) + 'cm'
    axarr.annotate(txt, (x[i],y[i]-10),size=15,color='b')

plt.show()
```

```python
output_dir = "./out/"
output_filename = "COT_VS_DRAG_ENERY.png"
if not os.path.exists(output_dir):
            os.makedirs(output_dir)
figure.savefig(output_dir + output_filename, dpi=450)
```
