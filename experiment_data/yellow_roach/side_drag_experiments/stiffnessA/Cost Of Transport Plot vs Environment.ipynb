```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
sns.set_palette("bright")
```

```python
figure, axarr = plt.subplots(1)

y = [113.5727159,48.5988870244,35,30.5484225944]
x = [4,6,8,10]

axarr.plot(x,y,color='b',marker='o')
axarr.axis([0,12,0,120])
axarr.set_ylabel('Cost of Transport')
axarr.set_xlabel('Gap Between Grass (cm)')

for i, txt in enumerate(x):
    txt = str(txt) + 'cm'
    axarr.annotate(txt, (x[i],y[i]),size=15,color='b')

plt.show()
```

```python
output_dir = "./out/"
output_filename = "CoT_Plot.png"
if not os.path.exists(output_dir):
            os.makedirs(output_dir)
figure.savefig(output_dir + output_filename, dpi=450)
```

```python
figure, axarr = plt.subplots(1)

y = [113.5727159,48.5988870244,35,30.5484225944]
x = [0.02555974843,0.05223650386,0.09112107623,0.06773333333]
n = [4,6,8,10]
axarr.plot(x,y,color='b',marker='o')
axarr.axis([0,0.1,0,120])
axarr.set_ylabel('Cost of Transport')
axarr.set_xlabel('Speed (m/s)')
axarr.text(0.068,110,u"\u2022 gap between grass",fontsize=15,color='b')
for i, txt in enumerate(n):
    txt = str(txt)+'cm'
    axarr.annotate(txt, (x[i],y[i]),size=15,color='b')

plt.show()
```

```python
output_dir = "./out/"
output_filename = "CoT_Plot_Speed.png"
if not os.path.exists(output_dir):
            os.makedirs(output_dir)
figure.savefig(output_dir + output_filename, dpi=450)
```
